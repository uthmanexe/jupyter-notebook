### Rainbow colors
ask for input of a favorite rainbow color first letter: ROYGBIV  

Using `if`, `elif`, and `else`:  
- print the color matching the letter  
    - R = Red  
    - O = Orange  
    - Y = Yellow  
    - G = Green
    - B = Blue
    - I = Indigo
    - V = Violet
    - else print "no match"


```python
# [ ] complete rainbow colors
choice = input("Enter the first letter of a color in the rainbow: ").strip().upper()

if choice == "R":
    print("Red")
elif choice("O"):
    print == "Orange"
elif choice("Y"):
    print == "Yellow"
elif choice("G"):
    print == "Green"
elif choice("B"):
    print == "Blue"
elif choice("I"):
    print == "Indigo"
elif choice("V"):
    print == "Violet"
else:
    print("No match")
```

```python
# [ ] make the code above into a function rainbow_color() that has a string parameter,
def rainbow_color(choice):
    if choice.startswith("R"):
        print("Red")
    elif choice.startswith("O"):
        print == "Orange"
    elif choice.startswith("Y"):
        print == "Yellow"
    elif choice.startswith("G"):
        print == "Green"
    elif choice.startswith("B"):
        print == "Blue"
    elif choice.startswith("I"):
        print == "Indigo"
    elif choice.startswith("V"):
        print == "Violet"
    else:
        print("No match")


# get input and call the function and return the matching color as a string or "no match" message.
choice = input(
    "Enter the first letter of a color in the rainbow: ").strip().upper()
# Call the function and print the return string.
rainbow_color(choice)
```

# &nbsp;  
**Create function age_20() that adds or subtracts 20 from your age for a return value based on current age** (use `if`) 
- call the funtion with user input and then use the return value in a sentence  
example `age_20(25)` returns **5**: 
> "5 years old, 20 years difference from now"

```python
# [ ] complete age_20()

age = int(input("Enter you age: ").strip())


def age_20(age):
    if age > 20:
        return f"{age - 20} years old, 20 years difference from now"
    else:
        return f"{age + 20} years old, 20 years from now"


age_20(age)
```

**create a function rainbow_or_age that takes a string argument**
- if argument is a digit return the value of calling age_20() with the str value cast as **`int`** 
- if argument is an alphabetical character return the value of calling rainbow_color() with the str
- if neither return FALSE

```python
# [ ]  create rainbow_or_age()
def rainbow_or_age(val):
    if val.isalpha():
        val_upper = val.upper()

        if val_upper.startswith("R"):
            return "Red"
        elif val_upper.startswith("O"):
            return "Orange"
        elif val_upper.startswith("Y"):
            return "Yellow"
        elif val_upper.startswith("G"):
            return "Green"
        elif val_upper.startswith("B"):
            return "Blue"
        elif val_upper.startswith("I"):
            return "Indigo"
        elif val_upper.startswith("V"):
            return "Violet"
        else:
            return "No match"

    elif val.isdigit():
        age = int(val)
        if age > 20:
            return f"{age} years old, {age - 20} years difference from 20"
        else:
            return f"{age} years old, {age + 20} years from now"

    else:
        return False


user_input = input("Enter a color starting letter or your age: ").strip()
print(rainbow_or_age(user_input))
```

```python
# [ ] Divide a larger number by a smaller number and print the integer part of the result
# don't divide by zero! if a zero is input make the result zero
# [ ] cast the answer to an integer to cut off the decimals and print the result

num1 = float(input().strip())
num2 = float(input().strip())

if num1 == 0 or num2 == 0:
    result = 0
elif num1 > num2:
    result = int(num1 / num2)
else:
    result = int(num2 / num1)

print(result)
```
