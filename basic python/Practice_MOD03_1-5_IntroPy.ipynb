### Rainbow colors
ask for input of a favorite rainbow color first letter: ROYGBIV  

Using `if`, `elif`, and `else`:  
- print the color matching the letter  
    - R = Red  
    - O = Orange  
    - Y = Yellow  
    - G = Green
    - B = Blue
    - I = Indigo
    - V = Violet
    - else print "no match"


```python
# [ ] complete rainbow colors
choice = input("Enter the first letter of a color in the rainbow: ").strip().upper()

if choice == "R":
    print("Red")
elif choice("O"):
    print == "Orange"
elif choice("Y"):
    print == "Yellow"
elif choice("G"):
    print == "Green"
elif choice("B"):
    print == "Blue"
elif choice("I"):
    print == "Indigo"
elif choice("V"):
    print == "Violet"
else:
    print("No match")
```

```python
# [ ] make the code above into a function rainbow_color() that has a string parameter, 
def rainbow_color(choice):
    if choice == "R":
        print("Red")
    elif choice("O"):
        print == "Orange"
    elif choice("Y"):
        print == "Yellow"
    elif choice("G"):
        print == "Green"
    elif choice("B"):
        print == "Blue"
    elif choice("I"):
        print == "Indigo"
    elif choice("V"):
        print == "Violet"
    else:
        print("No match")


# get input and call the function and return the matching color as a string or "no match" message.
choice = input(
    "Enter the first letter of a color in the rainbow: ").strip().upper()
# Call the function and print the return string.
rainbow_color(choice)
```
